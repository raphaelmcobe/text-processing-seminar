```python
text = """
Ground control to Major Tom (10, 9, 8, 7).
Commencing countdown, engines on (6, 5, 4, 3).
Check ignition, and may God's love be with you (2, 1, lift off).
"""
```

```python
import re
```

```python
text = re.sub("[^a-zA-Z0-9]",' ',text)
```

```python
text = re.sub('[0-9]+', '', text)
```

```python
text = re.sub('\s+', ' ', text)
```

```python
text = re.sub('^\s', '', text)
text = re.sub('\s$', '', text)
```

```python
print(text)
```

```python
# importing word_tokenize from nltk
from nltk.tokenize import word_tokenize
```

```python
token = word_tokenize(text)
print(token)
```

```python
# finding the frequency distinct in the tokens
# Importing FreqDist library from nltk and passing token into FreqDist
from nltk.probability import FreqDist
fdist = FreqDist(token)
fdist
```

## Movie lines datasets

```python
lines = pd.read_csv('movie_lines.txt', sep = '\+\+\+\$\+\+\+', engine = 'python', header=None, dtype=str)
```

```python
lines.head()
```

```python
lines = lines.iloc[:,[2,4]]
```

```python
lines.columns=['movie_id', 'lines']
```

```python
lines.head()
```

```python
lines = lines.fillna('').groupby('movie_id').agg({'lines': ' '.join}).reset_index()
```

```python
lines.movie_id = lines['movie_id'].str.strip()
```

```python
lines.iloc[0,0]
```

```python
movie_metadata = pd.read_csv('movie_titles_metadata.txt', sep = '\+\+\+\$\+\+\+', engine = 'python', header=None, dtype=str)
```

```python
movie_metadata.head()
```

```python
movie_metadata = movie_metadata.iloc[:,[0,1,2,5]]
```

```python
movie_metadata.columns=['movie_id', 'movie_name', 'year', 'genres']
```

```python
movie_metadata.movie_id = movie_metadata['movie_id'].str.strip()
movie_metadata.movie_name = movie_metadata['movie_name'].str.strip()
```

```python
list(movie_metadata.iloc[:,1])
```

```python
lines.shape
```

```python
movie_db = pd.merge(lines, movie_metadata, how='inner', on='movie_id')
```

```python
movie_db.head()
```

```python
ghostsbusters_lines = movie_db[movie_db['movie_name']=='ghostbusters'].iloc[0,1]
```

```python
movie_db.to_csv("movie_lines_db.csv", header=True, index=False)
```

```python
# importing stopwors from nltk library
from nltk import word_tokenize
from nltk.corpus import stopwords
english_stopwords = set(stopwords.words('english'))
```

```python
ghostbusters_token = word_tokenize(ghostsbusters_lines.lower())
```

```python
clean_lines = [word for word in ghostbusters_token if word not in english_stopwords] 
```

```python
print(clean_lines)
```

```python
# Importing Porterstemmer from nltk library
# Checking for the word ‘giving’ 
from nltk.stem import PorterStemmer
pst = PorterStemmer()
pst.stem("aiting")
```

```python
print([pst.stem(word) for word in clean_lines])
```
